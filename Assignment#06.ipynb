```python
class Car():
    def __init__(self, name, color, model, price):
        self.name = name
        self.color = color
        self.model = model
        self.price = price
        self.condition = 'New'
    def updateprice(self,newprice):
        self.price = newprice
    def updatecondition(self, newcondition):
        self.condition = newcondition
    def updatecolor(self, newcolor):
        self.color = newcolor
        


        
car1 = Car('Carola', 'black', 2007,'12 lac')
print(car1.name)
print(car1.color)
print(car1.model)
print(car1.price)
print(car1.condition)
print('\n')

car2 = Car ('BMW', 'black', 2018, '12 million')
print(car2.name)
print(car2.color, 'old color')
car2.updatecolor('white')
print(car2.color, 'new color')
print(car2.model)
print(car2.condition, 'old condition')
car2.updatecondition('used')
print(car2.condition, 'updated ')
print('\n')

car3 = Car('city', 'blue', 2019, '30 lac')
print(car3.name)
print(car3.color, 'old color')
car3.updatecolor('red')
print(car3.color, 'new color')
print(car3.model)
print(car3.price,'old price')
car3.updateprice('25 lac')
print(car3.price, 'new price')
print(car3.condition, 'old condition')
car3.updatecondition('used')
print(car3.condition,'condition updated')
print('\n')

car4 = Car('ferrari', 'yellow', 2019, '$330,050')
print(car4.name)
print(car4.color)
print(car4.model)
print(car4.price,'old price')
car4.updateprice('$349.050')
print(car4.price, 'new price')
print(car4.condition)
print('\n')

car5 = Car('land rover', 'black', 2018, '$80K')
print(car5.name)
print(car5.color, 'old color')
car5.updatecolor('green')
print(car5.color, 'new color')
print(car5.model)
print(car5.price, 'old price')
car5.updateprice("$100K")
print(car5.price, 'new price')
print(car5.condition)
print('\n')

print(car1)
print(car2)
print(car3)
print(car4)
print(car5)








```
